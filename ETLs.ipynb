```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_data.csv')
df
```

```python
df['total_price'] = df['quantity'] * df['price']
df
```

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
df
```

```python
# Generamos IDs únicos para cada dimensión

df['product_id'] = pd.factorize(df['category'] + df['price'].astype(str))[0] + 1
df['payment_method_id'] = pd.factorize(df['payment_method'])[0] + 1
df['shopping_mall_id'] = pd.factorize(df['shopping_mall'])[0] + 1
df['invoice_date_id'] = pd.factorize(df['invoice_date'].dt.strftime('%Y-%m-%d'))[0] + 1
df
```

```python

invoice = df[['invoice_no', 'customer_id', 'product_id', 'payment_method_id',
              'invoice_date_id', 'shopping_mall_id', 'quantity', 'total_price']]
invoice = invoice.rename(columns={'invoice_no': 'invoice_id'})
invoice
```

```python
dim_customer = df[['customer_id', 'gender', 'age']].drop_duplicates()
dim_customer['name'] = ''
dim_customer 
```

```python
dim_product = df[['product_id', 'category', 'price']].drop_duplicates()
dim_product['name'] = ''
dim_product
```

```python
dim_payment_method = df[['payment_method_id', 'payment_method']].drop_duplicates()
dim_payment_method = dim_payment_method.rename(columns={'payment_method': 'type'})
dim_payment_method['value_received'] = 0.0
dim_payment_method['value_returned'] = 0.0
dim_payment_method
```

```python
dim_date = df[['invoice_date_id', 'invoice_date']].drop_duplicates()
dim_date['day'] = dim_date['invoice_date'].dt.day
dim_date['month'] = dim_date['invoice_date'].dt.month
dim_date['year'] = dim_date['invoice_date'].dt.year
dim_date = dim_date.drop(columns=['invoice_date'])
dim_date
```

```python
dim_shopping_mall = df[['shopping_mall_id', 'shopping_mall']].drop_duplicates()
dim_shopping_mall = dim_shopping_mall.rename(columns={'shopping_mall': 'name'})
dim_shopping_mall
```
